```python
import random

# Function to roll a six-sided die
def roll_dice():
    return random.randint(1, 6)

# Function to handle a player's turn
def player_turn(player):
    score = 0
    print(f"Player {player}'s turn:")
    while True:
        roll = roll_dice()  # Roll the die
        print(f"You rolled: {roll}")  # Print the result of the roll
        if roll == 1:
            print("Oops! You rolled a 1 and lost your turn.")  # Inform the player they rolled a 1
            return 0  # End the turn with a score of 0
        score += roll  # Add the roll to the player's score
        print(f"Current score: {score}")  # Display the current score
        choice = input("Roll again? (y/n): ")  # Ask the player if they want to roll again
        if choice.lower() != 'y':  # If the player doesn't want to roll again
            break  # End the turn
    return score  # Return the final score for the turn

# Function to play the Pig game
def main():
    print("Welcome to the Pig Game!")
    num_players = int(input("Enter the number of players (2-4): "))  # Ask the user for the number of players
    target_score = 50  # Set the target score to win the game
    player_scores = [0] * num_players  # Initialize scores for all players to 0
    current_player = 0  # Start with the first player

    # Loop to play the game
    while True:
        # Take a turn for the current player and update their score
        player_scores[current_player] += player_turn(current_player + 1)
        print(f"Player {current_player + 1}'s total score: {player_scores[current_player]}")
        
        # Check if the current player has reached the target score
        if player_scores[current_player] >= target_score:
            print(f"Player {current_player + 1} wins!")
            break  # End the game if a player wins
        
        # Switch to the next player for the next turn
        current_player = (current_player + 1) % num_players

    # Determine the winner(s) based on the highest score
    max_score = max(player_scores)  # Find the highest score
    winners = [i + 1 for i, score in enumerate(player_scores) if score == max_score]  # Find the player(s) with the highest score
    print(f"Player(s) {', '.join(map(str, winners))} wins with a score of {max_score}!")  # Print the winner(s)

if __name__ == "__main__":
    main()
```
